```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#importing the Iris dataset with pandas
dataset = pd.read_csv('D:\INEI\Tesis\WEKA\Datos Convertidos\AcademicoAlumnoSocial.csv')
x = dataset.iloc[:, [4, 5, 6, 7, 8, 10, 11, 12, 13, 14, 15, 17, 18, 21, 23, 24, 25, 26, 27, 28, 29, 31, 33, 35]].values
dataset.describe()
```

```python
#Finding the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('Método del codo')
plt.xlabel('Número de Agrupaciones')
plt.ylabel('Suma de error cuadratico') #within cluster sum of squares
plt.show()
```
